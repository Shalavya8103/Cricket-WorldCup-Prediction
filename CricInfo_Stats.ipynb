```python
import pandas as pd
import numpy as np
```

```python
bat_stats=pd.read_html('https://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;orderby=won;team=1;team=15;team=2;team=25;team=3;team=40;team=5;team=6;team=7;team=8;template=results;type=team')[2]
bat_sr_stat=pd.read_html('https://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;groupby=team;orderby=runs;spanmin1=27+Sep+2020;spanval1=span;team=1;team=15;team=2;team=25;team=3;team=40;team=5;team=6;team=7;team=8;template=results;type=batting')[2]
bat_sr_stat=bat_sr_stat.rename(columns={'SR':'Bat_SR'})
ball_sr_stat=pd.read_html('https://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;groupby=team;orderby=wickets;spanmin1=27+Sep+2020;spanval1=span;team=1;team=15;team=2;team=25;team=3;team=40;team=5;team=6;team=7;team=8;template=results;type=bowling')[2]
ball_sr_stat=ball_sr_stat.rename(columns={'SR':'Ball_SR'})
fielding_stats=pd.read_html('https://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;groupby=team;orderby=dismissals;spanmin1=27+Sep+2020;spanval1=span;team=1;team=15;team=2;team=25;team=3;team=40;team=5;team=6;team=7;team=8;template=results;type=fielding')[2]
```

```python
stats=pd.DataFrame()
stats['Team']=bat_sr_stat['Team']
```

```python
#adding batting and fielding stats
stats['MatPlayed']=np.zeros(10)
stats['High_inns_score']=np.zeros(10)
stats['NO']=np.zeros(10)
stats['BatAvg']=np.zeros(10)
stats['Bat_SR']=np.zeros(10)
stats['Runs/Inns']=np.zeros(10)
stats['BF/Inns']=np.zeros(10)
stats['Hundreds/Inns']=np.zeros(10)
stats['Fifties/Inns']=np.zeros(10)
stats['Boundries/Inns']=np.zeros(10)
stats['Sixes/Inns']=np.zeros(10)
stats['FieldingDiss/Inns']=np.zeros(10)
stats['Stumping']=np.zeros(10)


for i in range (0,10):
  for j in range(0,10):
    if stats['Team'][i]==bat_stats['Team'][j]:
      stats['MatPlayed'][i]=bat_stats['Mat'][j]
      stats['High_inns_score'][i]=bat_stats['HS'][j]

    if stats['Team'][i]==bat_sr_stat['Team'][j]:
      stats['NO'][i]=bat_sr_stat['NO'][j]
      stats['BatAvg'][i]=bat_sr_stat['Ave'][j]
      stats['Bat_SR'][i]=bat_sr_stat['Bat_SR'][j]
      stats['BF/Inns']=bat_sr_stat['BF'][j]/bat_sr_stat['Mat'][j]
      stats['Runs/Inns'][i]=bat_sr_stat['Runs'][j]/bat_sr_stat['Mat'][j]
      stats['Hundreds/Inns'][i]=bat_sr_stat['100'][j]/bat_sr_stat['Mat'][j]
      stats['Fifties/Inns'][i]=bat_sr_stat['50'][j]/bat_sr_stat['Mat'][j]
      stats['Boundries/Inns'][i]=bat_sr_stat['4s'][j]/bat_sr_stat['Mat'][j]
      stats['Sixes/Inns'][i]=bat_sr_stat['6s'][j]/bat_sr_stat['Mat'][j]

    if stats['Team'][i]==fielding_stats['Team'][j]:
      stats['Stumping'][i]= fielding_stats['St'][j]
      stats['FieldingDiss/Inns'][i]= fielding_stats['D/I'][j]

```

```python
#adding bowling stats
stats['BallAvg']=np.zeros(10)
stats['Ball_SR']=np.zeros(10)
stats['Econ']=np.zeros(10)
stats['RunsConceded/Inns']=np.zeros(10)
stats['WickTaken/Inns']=np.zeros(10)

for i in range (0,10):
  for j in range(0,10):
    if stats['Team'][i]==ball_sr_stat['Team'][j]:
      stats['BallAvg'][i]=ball_sr_stat['Ave'][j]
      stats['Ball_SR'][i]=ball_sr_stat['Ball_SR'][j]
      stats['Econ'][i]=ball_sr_stat['Econ'][j]
      stats['RunsConceded/Inns'][i]=ball_sr_stat['Runs'][j]/ball_sr_stat['Mat'][j]
      stats['WickTaken/Inns'][i]=ball_sr_stat['Wkts'][j]/ball_sr_stat['Mat'][j]
```
